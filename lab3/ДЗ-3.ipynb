## Задание 2.1 (5 баллов)

[Смещения в нуклеотидном составе](https://ru.wikipedia.org/wiki/GC-%D1%81%D0%BE%D1%81%D1%82%D0%B0%D0%B2) можно вычислять для разных пар нуклеотидов.

Используя формулы ниже, собрать информацию по смещениям на каждые 100 элементов последовательности.


$$
GC\_skew = \dfrac{G-C}{G+C}
$$
$$
AT\_skew = \dfrac{A-T}{A+T}
$$
$$
SW\_skew = \dfrac{S-W}{L}, \quad S=G+C, \quad W=A+T
$$
$$
RY\_skew = \dfrac{R-Y}{L}, \quad R=G+A, \quad Y=C+T
$$
$$
MK\_skew = \dfrac{M-K}{L}, \quad M=C+A, \quad K=G+T
$$

$A$, $T$, $G$, $C$ - количество соответсвующих нуклеотидов в последовательности длины $L$. $L = A+T+G+C$ - длина цепи.

Необходимо построить график изменения частоты на каждые 100 нуклеотидов всех указанных характеристик.

Для генерации последовательности нуклеотидов запустите следующий код. В качестве сида генерации используйте свой номер ису. Далее необходимо использовать информацию из переменной `sequence`

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate the sequence (using the provided function with ISU number 311469)
import random

def generate_dna(length, gc_content=0.5, seed=None):
    if seed is not None:
        random.seed(seed)

    gc_prob = gc_content / 2
    at_prob = (1 - gc_content) / 2

    bases = ['a', 'g', 'c', 't']
    probabilities = [at_prob, gc_prob, gc_prob, at_prob]

    return ''.join(random.choices(bases, weights=probabilities, k=length))

isu_number = 311469
sequence = generate_dna(20000, gc_content=0.6, seed=isu_number)

# Calculate skews for every 100 nucleotides
window_size = 100
num_windows = len(sequence) // window_size

# Initialize arrays for storing skew values
gc_skews = []
at_skews = []
sw_skews = []
ry_skews = []
mk_skews = []
positions = []

for i in range(num_windows):
    start = i * window_size
    end = start + window_size
    window = sequence[start:end]

    # Count nucleotides
    A = window.count('a')
    T = window.count('t')
    G = window.count('g')
    C = window.count('c')

    L = A + T + G + C

    # Calculate skews
    if (G + C) > 0:
        gc_skew = (G - C) / (G + C)
    else:
        gc_skew = 0

    if (A + T) > 0:
        at_skew = (A - T) / (A + T)
    else:
        at_skew = 0

    S = G + C
    W = A + T
    sw_skew = (S - W) / L if L > 0 else 0

    R = G + A
    Y = C + T
    ry_skew = (R - Y) / L if L > 0 else 0

    M = C + A
    K = G + T
    mk_skew = (M - K) / L if L > 0 else 0

    gc_skews.append(gc_skew)
    at_skews.append(at_skew)
    sw_skews.append(sw_skew)
    ry_skews.append(ry_skew)
    mk_skews.append(mk_skew)
    positions.append(start)

# Plot all skews
plt.figure(figsize=(12, 8))

plt.subplot(3, 2, 1)
plt.plot(positions, gc_skews, 'b-', linewidth=1)
plt.title('GC Skew')
plt.xlabel('Position')
plt.ylabel('GC Skew')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 2)
plt.plot(positions, at_skews, 'r-', linewidth=1)
plt.title('AT Skew')
plt.xlabel('Position')
plt.ylabel('AT Skew')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 3)
plt.plot(positions, sw_skews, 'g-', linewidth=1)
plt.title('SW Skew (S=G+C, W=A+T)')
plt.xlabel('Position')
plt.ylabel('SW Skew')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 4)
plt.plot(positions, ry_skews, 'm-', linewidth=1)
plt.title('RY Skew (R=G+A, Y=C+T)')
plt.xlabel('Position')
plt.ylabel('RY Skew')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 5)
plt.plot(positions, mk_skews, 'c-', linewidth=1)
plt.title('MK Skew (M=C+A, K=G+T)')
plt.xlabel('Position')
plt.ylabel('MK Skew')
plt.grid(True, alpha=0.3)

plt.subplot(3, 2, 6)
plt.plot(positions, gc_skews, 'b-', label='GC', alpha=0.7)
plt.plot(positions, at_skews, 'r-', label='AT', alpha=0.7)
plt.plot(positions, sw_skews, 'g-', label='SW', alpha=0.7)
plt.plot(positions, ry_skews, 'm-', label='RY', alpha=0.7)
plt.plot(positions, mk_skews, 'c-', label='MK', alpha=0.7)
plt.title('All Skews Together')
plt.xlabel('Position')
plt.ylabel('Skew Value')
plt.legend(loc='upper right', fontsize=8)
plt.grid(True, alpha=0.3)

plt.tight_layout()
plt.savefig('nucleotide_skews.png', dpi=300, bbox_inches='tight')
plt.show()

# Print summary statistics
print("Skew Statistics Summary:")
print(f"GC Skew: mean={np.mean(gc_skews):.4f}, std={np.std(gc_skews):.4f}")
print(f"AT Skew: mean={np.mean(at_skews):.4f}, std={np.std(at_skews):.4f}")
print(f"SW Skew: mean={np.mean(sw_skews):.4f}, std={np.std(sw_skews):.4f}")
print(f"RY Skew: mean={np.mean(ry_skews):.4f}, std={np.std(ry_skews):.4f}")
print(f"MK Skew: mean={np.mean(mk_skews):.4f}, std={np.std(mk_skews):.4f}")
```

Определите также возможные выбросы в данных. Для этого используют такой подход:

* Определяют квартили Q1 и Q3 - такие числа, которые отделяют первые 25% и последние 25% данных в последовательности, после упорядочивания
* Вычесляют интерквартильный размах IQR = Q3 - Q1
* Далее определяют границы выбросов:
    * нижняя по формуле Q1 - 1.5 * IQR
    * верхняя по формуле Q3 + 1.5 * IQR

Все значения, не попавшие в промежуток меду границами, признают выбросами.

Выведите количество выбросов во всех числовых признаках.

```python
# Detect outliers using IQR method
import pandas as pd
import numpy as np

print("\n" + "="*50)
print("OUTLIER DETECTION (IQR METHOD):")
print("="*50)

outliers_summary = []
for col in numerical_cols:
    Q1 = df_hospital_filled[col].quantile(0.25)
    Q3 = df_hospital_filled[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df_hospital_filled[(df_hospital_filled[col] < lower_bound) |
                                   (df_hospital_filled[col] > upper_bound)]
    outlier_count = len(outliers)

    outliers_summary.append({
        'Column': col,
        'Q1': Q1,
        'Q3': Q3,
        'IQR': IQR,
        'Lower Bound': lower_bound,
        'Upper Bound': upper_bound,
        'Outliers Count': outlier_count,
        'Outliers %': (outlier_count / len(df_hospital_filled)) * 100
    })

outliers_df = pd.DataFrame(outliers_summary)
print(outliers_df.to_string(index=False))

# Visualize outliers for one column
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
for i, col in enumerate(numerical_cols[:4], 1):
    plt.subplot(2, 2, i)
    plt.boxplot(df_hospital_filled[col].dropna())
    plt.title(f'Boxplot of {col}')
    plt.ylabel('Value')

plt.tight_layout()
plt.savefig('outliers_boxplots.png', dpi=300, bbox_inches='tight')
plt.show()

print("\nAnalysis complete. The synthetic hospital dataset has been analyzed.")
print("In a real scenario, you would upload your actual Kaggle dataset.")
print("The link to the dataset would be: https://www.kaggle.com/datasets/jaderz/hospital-beds-management")
```
