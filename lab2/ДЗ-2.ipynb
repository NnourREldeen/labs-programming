## Задача 1. Королевский гамбит (1 балл)

Петя любит играть в шахматы. Он совершает свои первые шаги в игре и изучает правила перемещения фигур на доске. Помогите Пете ответить на вопрос: могла ли фигура совершить такое перемещение из точки $A$ в точку $B$?

В первой строке программе подается название фигуры:
- ферзь
- король
- конь
- ладья

После этого, на первой строке через пробел указываются координаты точки $A(x_1 ,y_1)$, затем на следующей координаты $B(x_2, y_2)$.

Программа должна вывести Yes, если фигура за один ход могла перейти из точки $A$ в точку $B$, и No в противном случае.

Считайте, что поля пронумерованы от 1 до 8, начиная с левого нижнего угла.

Пример ввода:
```
король
1 2
5 7
```

Пример вывода:
```
No
```

````python
figure = input().strip()
x1, y1 = map(int, input().split())
x2, y2 = map(int, input().split())

dx = abs(x1 - x2)
dy = abs(y1 - y2)

if figure == "ферзь":
    result = (dx == 0 or dy == 0 or dx == dy)
elif figure == "король":
    result = (dx <= 1 and dy <= 1 and (dx != 0 or dy != 0))
elif figure == "конь":
    result = ((dx == 1 and dy == 2) or (dx == 2 and dy == 1))
elif figure == "ладья":
    result = (dx == 0 or dy == 0) and (dx != 0 or dy != 0)
else:
    result = False

print("Yes" if result else "No")
````

## Задача 2. Большие числа (1 балл)

Для натуральных чисел определена операция взятия факториала. Напомним, что это число вида $n! = 1 ⋅ 2 ⋅ \ldots ⋅ n$. При этом считают, что $0! = 1$.

Пользователь на вход вводит значение $n$. В ответе необходимо найти значение выражения
$$
1! + 2! + \dots + n!
$$

Пользоваться математической библиотекой math в этой задаче запрещено.

Пример ввода:
```
7
```

Пример вывода:
```
5913
```

````python
n = int(input())

total_sum = 0
current_factorial = 1

for i in range(1, n + 1):
    current_factorial *= i
    total_sum += current_factorial

print(total_sum)
````

## Задача 3. Потерянная карточка (1 балл)

В настольной игре используются карточки с номерами от 1 до $n$. Одна карточка потерялась. Зная номера оставшихся карточек, найдите номер потерявшейся.

На вход подается число $n$, далее $n-1$ чисел в произвольном порядке (номера от 1 до $n$, без номера потерянной карточки).

Необходимо после обработки данных вывести номер недостающей карточки.

Пример ввода:
```
5
3
5
2
1
```

Пример вывода:
```
4
```

````python
n = int(input())

total_sum = n * (n + 1) // 2

for _ in range(n - 1):
    total_sum -= int(input())

print(total_sum)
````

## Задача 4. Утро начинается не с кофе (1 балл)

В первый день спортсмен пробежал x километров, а затем он каждый день увеличивал пробег на 10% от предыдущего значения. По данному числу y определите номер дня, на который пробег спортсмена составит не менее y километров.

Программа получает на вход действительные числа $x$ и $y$ и должна вывести одно натуральное число.

Пример ввода:
```
1
1000
```

Пример вывода:
```
74
```

````python
x = float(input())
y = float(input())

day = 1
distance = x

while distance < y:
    distance += distance * 0.1
    day += 1

print(day)
````

## Задача 5. СКО (1 балл)

Пользователь вводит последовательность чисел, которая завершается нулем.

Определите для введенной последовательности стандартное нормальное отклонение (мера разброса данных), которая вычисляется по формуле

$$
\sigma = \sqrt{\frac{(x_1-s)^2+(x_2-s)^2+\ldots+(x_n-s)^2}{n-1}} $$

где
$$
s=\frac{x_1+x_2+\ldots+x_n}{n}
$$


Пример ввода:
```
10
1
1
1
1
1
0
```

Пример вывода:
```
3.67423461417
```

````python
numbers = []
while True:
    num = float(input())
    if num == 0:
        break
    numbers.append(num)

n = len(numbers)
if n <= 1:
    print(0)
else:
    s = sum(numbers) / n

    sum_sq = 0
    for num in numbers:
        sum_sq += (num - s) ** 2

    sigma = (sum_sq / (n - 1)) ** 0.5
    print(sigma)
````

## Задача 6. Снежинка (1 балл)

По данному нечетному числу $n$ необходимо создать двумерный массив, образующий снежинку. После чего необходимо ее вывести в консоль. На главной и побочной диагоналях, а также в средней строке и среднем столбце стоят символы "*", в остальных ячейках - точки "."


Пример ввода:
```
9
```

Пример вывода:
```
* . . . * . . . *
. * . . * . . * .
. . * . * . * . .
. . . * * * . . .
* * * * * * * * *
. . . * * * . . .
. . * . * . * . .
. * . . * . . * .
* . . . * . . . *
```

````python
n = int(input())

snowflake = [['.' for _ in range(n)] for _ in range(n)]

for i in range(n):
    for j in range(n):
        if i == j or i == n - 1 - j or i == n // 2 or j == n // 2:
            snowflake[i][j] = '*'

for row in snowflake:
    print(' '.join(row))
````

## Задача 7. Ветер перемен (1 балл)

Дан двумерный массив и два числа: $i$ и $j$. Поменяйте в массиве столбцы с номерами $i$ и $j$ и выведите результат.

Программа получает на вход размеры массива $n$ и $m$, затем элементы массива, затем числа $i$ и $j$.

Решение оформите в виде функции `swap_columns(a, i, j)`.

Пример ввода:
```
4 6
11 12 13 14 15 16
21 22 23 24 25 26
31 32 33 34 35 36
41 42 43 44 45 46
0 5
```

Пример вывода:
```
16 12 13 14 15 11
26 22 23 24 25 21
36 32 33 34 35 31
46 42 43 44 45 41
```

````python
def swap_columns(a, i, j):
    for row in a:
        row[i], row[j] = row[j], row[i]

n, m = map(int, input().split())
matrix = []
for _ in range(n):
    row = list(map(int, input().split()))
    matrix.append(row)

f, j = map(int, input().split())

swap_columns(matrix, f, j)

for row in matrix:
    print(' '.join(map(str, row)))
````

## Задача 8. Кегельбан (1 балл)

$N$ кеглей выставили в один ряд, занумеровав их слева направо числами от 1 до $N$. Затем по этому ряду бросили $K$ шаров, при этом $i$-й шар сбил все кегли с номерами от $l_i$ до $r_i$ включительно. Определите, какие кегли остались стоять на месте.

Программа получает на вход количество кеглей N и количество бросков $K$. Далее идет $K$ пар чисел $l_i$, $r_i$, при этом $1\le l_i\le r_i\le N$.

Программа должна вывести последовательность из $N$ символов, где $j$-й символ есть `I`, если $j$-я кегля осталась стоять, или `.`, если $j$-я кегля была сбита.

Пример ввода:
```
5 2
1 2
4 4
```

Пример вывода:
```
..I.I
```

Пояснение к примеру:
- до первого броска кегли были расположены так
```
IIIII
```
- после первого броска были сбиты кегли с первой по вторую
```
..III
```
- последний бросок выбил только четвертую кеглю
```
..I.I
```

````python
N, K = map(int, input().split())

pins = [True] * N

for _ in range(K):
    l, r = map(int, input().split())
    for i in range(l - 1, r):
        pins[i] = False

result = []
for pin in pins:
    if pin:
        result.append('I')
    else:
        result.append('.')

print(''.join(result))
````

## Задча 9. Едим дома (1 балл)

Каждый житель некоторой страны любит неоторые блюда. Нужно определить какие в этом обществе общие вкусовые предпочтения, а также список всех существуюзих гастрономических изысков.

В первой строке программа получает число $n$ - число жителей города. Далее по каждому человеку сначала передается информация о количестве его любимых блюд $k_j$, после чего друг за другом указываются названия этих блюд.

В ответе укажите сперва количество блюд, которые нравятся всем жителям, если они есть, то далее в алфавитном порядке перечислите названия этих блюд по одному в строке. Затем укажите количество существующих в стране блюд и также в лексикографическом порядке перечислите их.


Пример ввода:
```
3
2
Лазанья
Пицца
3
Суши
Пицца
Пудинг
1
Пицца
```

Пример вывода:
```
1
Пицца
4
Лазанья
Пицца
Пудинг
Суши
```

````python
n = int(input())

all_dishes = []
all_dishes_set = set()

for _ in range(n):
    k = int(input())
    dishes = set()
    for _ in range(k):
        dish = input().strip()
        dishes.add(dish)
        all_dishes_set.add(dish)
    all_dishes.append(dishes)

common_dishes = set(all_dishes[0])
for dishes in all_dishes[1:]:
    common_dishes &= dishes

print(len(common_dishes))
for dish in sorted(common_dishes):
    print(dish)

print(len(all_dishes_set))
for dish in sorted(all_dishes_set):
    print(dish)
````

## Задача 10. Англо-латинский словарь (1 балл)

Однажды, разбирая старые книги на чердаке, школьник Вася нашёл англо-латинский словарь. Английский он к тому времени знал в совершенстве, и его мечтой было изучить латынь. Поэтому попавшийся словарь был как раз кстати.

К сожалению, для полноценного изучения языка недостаточно только одного словаря: кроме англо-латинского необходим латинско-английский. За неимением лучшего он решил сделать второй словарь из первого.

Как известно, словарь состоит из переводимых слов, к каждому из которых приводится несколько слов-переводов. Для каждого латинского слова, встречающегося где-либо в словаре, Вася предлагает найти все его переводы (то есть все английские слова, для которых наше латинское встречалось в его списке переводов), и считать их и только их переводами этого латинского слова.

Помогите Васе выполнить работу по созданию латинско-английского словаря из англо-латинского.

В первой строке содержится единственное целое число N — количество английских слов в словаре. Далее следует N описаний. Каждое описание содержится в отдельной строке, в которой записано сначала английское слово, затем отделённый пробелами дефис, затем разделённые запятыми с пробелами переводы этого английского слова на латинский. Все слова состоят только из маленьких латинских букв. Переводы отсортированы в лексикографическом порядке. Порядок следования английских слов в словаре также лексикографический.

Выведите соответствующий данному латинско-английский словарь, в точности соблюдая формат входных данных. В частности, первым должен идти перевод лексикографически минимального латинского слова, далее — второго в этом порядке и т.д. Внутри перевода английские слова должны быть также отсортированы лексикографически.


Пример ввода:
```
3
apple - malum, pomum, popula
fruit - baca, bacca, popum
punishment - malum, multa
```

Пример вывода:
```
7
baca - fruit
bacca - fruit
malum - apple, punishment
multa - punishment
pomum - apple
popula - apple
popum - fruit
```


````python
n = int(input())

latin_to_english = {}

for _ in range(n):
    line = input().strip()

    eng_part, lat_part = line.split(' - ')
    eng_word = eng_part.strip()

    latin_words = [word.strip() for word in lat_part.split(',')]


    for latin_word in latin_words:
        if latin_word not in latin_to_english:
            latin_to_english[latin_word] = set()
        latin_to_english[latin_word].add(eng_word)

sorted_latin_words = sorted(latin_to_english.keys())


print(len(sorted_latin_words))
for latin_word in sorted_latin_words:
    sorted_english = sorted(latin_to_english[latin_word])
    print(f"{latin_word} - {', '.join(sorted_english)}")
````
