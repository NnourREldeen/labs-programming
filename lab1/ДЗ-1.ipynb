### Работа со строками (2 балла)

Задание 2.1

Напишите программу, которая считывает строку и выводит следующую информацию (каждый пункт в новой строке):

1. Третий символ этой строки
2. Предпоследний символ этой строки
3. Первые пять символов этой строки
4. Всю строку, кроме последних трех символов
5. Все символы, расположенные на четных позизиях
6. Все символы в обратном порядке
7. Длину всей строки
8. Количество всех символов, расположенных на нечетных позициях

```python
s = input("Введите строку: ")

# 1. Третий символ
if len(s) >= 3:
    print(f"1. Третий символ: {s[2]}")
else:
    print("1. Строка слишком короткая")

# 2. Предпоследний символ
if len(s) >= 2:
    print(f"2. Предпоследний символ: {s[-2]}")
else:
    print("2. Строка слишком короткая")

# 3. Первые пять символов
print(f"3. Первые пять символов: {s[:5]}")

# 4. Вся строка, кроме последних трех символов
if len(s) >= 3:
    print(f"4. Строка без последних трех символов: {s[:-3]}")
else:
    print("4. Строка слишком короткая")

# 5. Символы на четных позициях (индексы 0, 2, 4...)
print(f"5. Символы на четных позициях: {s[::2]}")

# 6. Строка в обратном порядке
print(f"6. Строка в обратном порядке: {s[::-1]}")

# 7. Длина строки
print(f"7. Длина строки: {len(s)}")

# 8. Количество символов на нечетных позициях
odd_chars = s[1::2]
print(f"8. Количество символов на нечетных позициях: {len(odd_chars)}")
```

Задание 3.1

Напишите программу, которая меняет значения в переменных a, b c по следующему правилу:
* в переменную a поместить значение $a + 2b - 3c$
* в переменную b поместить значение $a^2 + b^2 - c^2$
* в переменную c поместить значение $\dfrac{a+b+c}{a^2+b^2+c^2+1}$

Вывести значение переменных  a, b, c

```python
a = float(input("Введите a: "))
b = float(input("Введите b: "))
c = float(input("Введите c: "))

# Сохраняем исходные значения
a_orig, b_orig, c_orig = a, b, c

# Вычисляем новые значения
a_new = a_orig + 2*b_orig - 3*c_orig
b_new = a_orig**2 + b_orig**2 - c_orig**2
c_new = (a_orig + b_orig + c_orig) / (a_orig**2 + b_orig**2 + c_orig**2 + 1)

print(f"Новое значение a: {a_new}")
print(f"Новое значение b: {b_new}")
print(f"Новое значение c: {c_new}")
```

Задача *.1 (2 балл)

Помогите Паше написать программу, которая будет искать по трем сторона площадь треугольника. Считайте, что на вход подаются числа, которые являются сторонами существующего треугольника. Для решения рекомендуется использовать формулу Герона
$$
S = \sqrt{p(p-a)(p-b)(p-c)}, \quad p = \dfrac{a+b+c}{2}
$$

```python
import math

a = float(input("Введите сторону a: "))
b = float(input("Введите сторону b: "))
c = float(input("Введите сторону c: "))

# Проверка на существование треугольника
if a + b > c and a + c > b and b + c > a:
    p = (a + b + c) / 2
    area = math.sqrt(p * (p - a) * (p - b) * (p - c))
    print(f"Площадь треугольника: {area:.2f}")
else:
    print("Треугольник с такими сторонами не существует")
```

Задача *.2 (2 балла)

Паша очень любит кататься на общественном транспорте, а получая билет, сразу проверяет, счастливый ли ему попался. Билет считается счастливым, если сумма первых трех цифр совпадает с суммой последних трех цифр номера билета.

Однако Паша очень плохо считает в уме, поэтому попросил вас написать программу, которая проверит равенство сумм и выведет "Счастливый", если суммы совпадают, и "Обычный", если суммы различны.

На вход программе подаётся строка из шести цифр.

Выводить нужно только слово "Счастливый" или "Обычный", с большой буквы.

```python
ticket = input("Введите номер билета (6 цифр): ")

if len(ticket) != 6 or not ticket.isdigit():
    print("Некорректный номер билета")
else:
    first_sum = int(ticket[0]) + int(ticket[1]) + int(ticket[2])
    second_sum = int(ticket[3]) + int(ticket[4]) + int(ticket[5])

    if first_sum == second_sum:
        print("Счастливый")
    else:
        print("Обычный")
```

Задача *.3 (2 балла)

[GC-состав](https://ru.wikipedia.org/wiki/GC-%D1%81%D0%BE%D1%81%D1%82%D0%B0%D0%B2) является важной характеристикой геномных последовательностей и определяется как процентное соотношение суммы всех гуанинов и цитозинов к общему числу нуклеиновых оснований в геномной последовательности.

Напишите программу, которая вычисляет процентное содержание символов G (гуанин) и C (цитозин) в введенной строке (программа **не должна зависеть** от регистра вводимых символов).

Например, в строке "acggtgttat" процентное содержание символов G и C равно $\dfrac{4}{10}\cdot 100 = 40.0$, где 4 - это количество символов G и C,  а 10 - это длина строки.

```python
sequence = input("Введите геномную последовательность: ").upper()

if sequence:  # проверка, что строка не пустая
    gc_count = sequence.count('G') + sequence.count('C')
    gc_percentage = (gc_count / len(sequence)) * 100
    print(f"GC-состав: {gc_percentage:.1f}%")
else:
    print("Введена пустая строка")
```

Задача *.4 (2 балла)

Напишите программу, которая выводит наибольний и наименьший элементсписка (можно для упрощения создать список внутри программы, без считывания). Решение не должно использовать встроенную функцию `max()` или `min()`.

```python
numbers = [42, 15, 73, 8, 19, 56, 91, 23, 67, 34, 5]

# Инициализируем переменные
max_num = numbers[0]
min_num = numbers[0]

# Ищем максимальный и минимальный элементы
for num in numbers:
    if num > max_num:
        max_num = num
    if num < min_num:
        min_num = num

print(f"Список: {numbers}")
print(f"Наибольший элемент: {max_num}")
print(f"Наименьший элемент: {min_num}")

# Альтернативный вариант с вводом списка от пользователя
print("\nАльтернативный вариант с вводом от пользователя:")
user_input = input("Введите числа через пробел: ")
user_numbers = [float(x) for x in user_input.split()]

if user_numbers:  # если список не пустой
    user_max = user_numbers[0]
    user_min = user_numbers[0]

    for num in user_numbers:
        if num > user_max:
            user_max = num
        if num < user_min:
            user_min = num

    print(f"Наибольший элемент: {user_max}")
    print(f"Наименьший элемент: {user_min}")
else:
    print("Список пустой")
```
